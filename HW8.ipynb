```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_excel("1.4_02.xlsx", engine='openpyxl')
df
```

```python
df1 = df[['Регіон','Рік', 'сільське господарство']]
df1.loc[:, 'Галузь'] = 'сільське господарство'
df1.columns = ['Регіон','Рік', 'Енергія', 'Галузь']


df2 = df[['Регіон','Рік', 'промисловість']]
df2.loc[:, 'Галузь'] = 'промисловість'
df2.columns = ['Регіон','Рік', 'Енергія', 'Галузь']


df3 = df[['Регіон','Рік', 'будівництво']]
df3.loc[:, 'Галузь'] = 'будівництво'
df3.columns = ['Регіон','Рік', 'Енергія', 'Галузь']


df4 = df[['Регіон','Рік', 'послуги']]
df4.loc[:, 'Галузь'] = 'послуги'
df4.columns = ['Регіон','Рік', 'Енергія', 'Галузь']


df5 = df[['Регіон','Рік', 'державне управління']]
df5.loc[:, 'Галузь'] = 'державне управління'
df5.columns = ['Регіон','Рік', 'Енергія', 'Галузь']


df_long = pd.concat([df1, df2, df3, df4, df5])
df_long['Рік']=df_long['Рік'].apply(lambda x: x.strip())
df_long
```

```python
alt.Chart(df_long[df_long['Рік']=='2010']).mark_point().encode(
    x = alt.X('Регіон:N'),
    y = alt.Y('Галузь:N'),
    size = alt.Size('Енергія'),
    tooltip = [
        alt.Tooltip('Регіон:N'),
        alt.Tooltip('Галузь:N'),
        alt.Tooltip('Енергія:N')
    ]
).properties(width = 950, height = 500, title = 'Розподіл енергії за галузями в Україні, 2010 рік')
```

```python
alt.Chart(df_long[df_long['Рік']=='2017']).mark_point().encode(
    x = alt.X('Регіон:N'),
    y = alt.Y('Галузь:N'),
    size = alt.Size('Енергія'),
    color = alt.Color(value='red'),
    tooltip = [
        alt.Tooltip('Регіон:N'),
        alt.Tooltip('Галузь:N'),
        alt.Tooltip('Енергія:N')
    ]
).properties(width = 950, height = 500, title = 'Розподіл енергії за галузями в Україні, 2017 рік')
```

Дані графіки, на відміну від оригінального набагато менш громіздкі, також інтерактивність полегшує читання даних. На мою думку, найкраще розділити такі дані на два графіки, щоб окремо побачити різницю між двома заданими роками.
